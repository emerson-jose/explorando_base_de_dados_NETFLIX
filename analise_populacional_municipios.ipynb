# Análise Populacional dos Municípios Brasileiros

```python
import pandas as pd

# Carregar os dados
url = 'https://raw.githubusercontent.com/alexvaroz/data_science_alem_do_basico/master/POP2021_20221212.csv'
df = pd.read_csv(url)

# Padronizar os nomes das colunas
df.columns = df.columns.str.strip().str.upper().str.replace(" ", "")

# Verificar nomes das colunas
print(df.columns)
df.head()
```

## 1) Quantos municípios possuem população maior que a média do país?

```python
# Carregar os dados (código movido da célula 3JW0NFn_3z8H)
url = 'https://raw.githubusercontent.com/alexvaroz/data_science_alem_do_basico/master/POP2021_20221212.csv'
df = pd.read_csv(url)

# Padronizar os nomes das colunas (código movido da célula 3JW0NFn_3z8H)
df.columns = df.columns.str.strip().str.upper().str.replace(" ", "")

media_populacao = df['POPULACAO_ESTIMADA'].mean()
municipios_acima_media = df[df['POPULACAO_ESTIMADA'] > media_populacao]
print(f"Municípios com população acima da média: {len(municipios_acima_media)}")
```

## 2) Quais os 5 Municípios menos populosos?

```python
df.sort_values('POPULACAO_ESTIMADA').head(5)[['NOME_MUNICIPIO', 'UF', 'POPULACAO_ESTIMADA']]
```

## 3) Quais os 10 Municípios mais populosos?

```python
df.sort_values('POPULACAO_ESTIMADA', ascending=False).head(10)[['NOME_MUNICIPIO', 'UF', 'POPULACAO_ESTIMADA']]
```

## 4) Qual a média da população por estado?

```python
df.groupby('UF')['POPULACAO_ESTIMADA'].mean().sort_values(ascending=False)
```

## 5) Qual a média da população por região?

```python
df.groupby('NOME_MUNICIPIO')['POPULACAO_ESTIMADA'].mean().sort_values(ascending=False)
```

## 6) Quantos municípios existem em cada estado?

```python
df['UF'].value_counts().sort_values(ascending=False)
```

## 7) Quantidade de municípios por região

```python
df['NOME_MUNICIPIO'].value_counts().sort_values(ascending=False)
```

## 8) Quais estados possuem "São Domingos" como município?

```python
df[df['NOME_MUNICIPIO'].str.upper() == 'SÃO DOMINGOS']['UF'].unique()
```

## 9) Quais são os 10 nomes de município mais comuns?

```python
df['NOME_MUNICIPIO'].value_counts().head(10)
```

## 10) Diferença entre a maior e menor população de município por UF

```python
df.groupby('UF')['POPULACAO_ESTIMADA'].agg(lambda x: x.max() - x.min()).sort_values(ascending=False)
```
